```python
import os
import numpy as np
import pandas as pd
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
```

```python
grid_step = 1.0
lat_min, lat_max = -10, 30
lon_min, lon_max = -30, 20
lat_array = np.arange(lat_min, lat_max+1, grid_step)
lon_array = np.arange(lon_min, lon_max+1, grid_step)

fig, ax = plt.subplots(figsize=(16, 10), subplot_kw={'projection': ccrs.PlateCarree()}, tight_layout=True)
# # Field plot
# im = da_2d.plot.pcolormesh(
#     ax=ax,
#     x='longitude',
#     y='latitude',
#     transform=ccrs.PlateCarree(),
#     cmap='Blues',
#     add_colorbar=False
# )

# # Colorbar
# cbar = plt.colorbar(im, ax=ax, orientation='horizontal', pad=0.08, shrink=0.7)
# cbar.set_label(f"Precipitazione Totale ({da_2d.units})", fontsize=11)
ax.set_extent([lon_min, lon_max, lat_min, lat_max], crs=ccrs.PlateCarree())
#  Elements of the map
ax.add_feature(cfeature.LAND, facecolor='lightgray', zorder=0)
ax.add_feature(cfeature.OCEAN, facecolor='aliceblue', zorder=0)
ax.add_feature(cfeature.COASTLINE, linewidth=1.2, edgecolor='black', zorder=2)
ax.add_feature(cfeature.BORDERS, linewidth=0.8, edgecolor='crimson', linestyle='--', zorder=2)


ax.set_xticks(lon_array, crs=ccrs.PlateCarree())
ax.set_yticks(lat_array, crs=ccrs.PlateCarree())
ax.xaxis.set_major_formatter(LONGITUDE_FORMATTER)
ax.yaxis.set_major_formatter(LATITUDE_FORMATTER)
ax.tick_params(axis='x', labelsize=7, labelrotation=90)
ax.tick_params(axis='y', labelsize=7)

ax.gridlines(draw_labels=False, linewidth=1, color='gray', alpha=0.4, linestyle=':')
ax.set_title(f"Domain", fontsize=14, pad=15)

plt.show()
```
